Importe de bibliotecas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import Image
from matplotlib import pyplot as plt
%matplotlib inline
sns.set(style="darkgrid")
```

Leitura de arquivo de metadados

```python
df = pd.read_excel('input_UFSC_21_01_2022_v1.xlsx')
print(df)
```

Limpeza de colunas (deixando apenas as que serão usadas para checagem, não as com detalhes específicos)

```python
df1 = df[['AGENTE','STATION NAME','idName','XML_BSB','PDC_BSB','Busca_MedPlot_BSB','XML_RJ','PDC_RJ','Busca_MedPlot_RJ']]
print(df1)
```

Troca de valores para facilitar leitura

```python
df1= df1.replace([0], 'ausente')
df1= df1.replace([1], 'presente')
df1['STATION NAME'] = df['STATION NAME'].fillna('Sem_Informações')
df1['idName'] = df1['idName'].fillna('Sem_Informações')
df1['XML_RJ'] = df1['XML_RJ'].fillna('Sem_Informações')
```

```python
print(df1)
```

# Confecção de Gráficos e sinalização de faltantes

```python
agentes= np.unique(df1['AGENTE'])

print(agentes)
print('O total de agentes é', len(agentes))
```

# *BSB*

# Análise de fluxos presentes na planilha de metadados x presentes no XML.
# Aqui são sinalizadas as PMUs que não constam no XML (medplot)

```python
total=[]
ausentes=[]
presentes=[]
fluxosausentes=[]

for x in range(len(agentes)):
    totalparcial=[]
    ausentesparcial=[]
    for l in range(len(df1['AGENTE'])):
        if df1['AGENTE'][l] == agentes[x]:
            totalparcial.append('1')
            if df1['XML_BSB'][l] == 'ausente':
                ausentesparcial.append('1')
                texto= df1['AGENTE'][l] + ' ' + df1['idName'][l]
                fluxosausentes.append(texto)
    total.append(len(totalparcial))
    ausentes.append(len(ausentesparcial))
    presentes.append(len(totalparcial) - len(ausentesparcial))

resumo = [agentes,presentes,ausentes,total]

planilha = pd.DataFrame (resumo).transpose()
planilha.columns = ['Agente', 'Presentes', 'Ausentes', 'Total']

plt.figure(figsize=(20,14))
plt.title('Metadados x XML')
sns.countplot(y="AGENTE", hue="XML_BSB", edgecolor="0.2", data=df1)
plt.savefig('Metadados_XML_BSB.png', format='png')
plt.show()

print('\n#####################################################')
print (planilha)

print('\n#####################################################\nOs fluxos ausentes para esta análise são: \n')
for x in range(len(fluxosausentes)):
    print(fluxosausentes[x])
```

# Resumo de análises feitas no servidor RJ

```python
faltanteXML=[[],[],[]]
for x in range(len(df1['PDC_RJ'])):
    if df1['PDC_RJ'][x]=='presente':
        if df1['XML_RJ'][x]=='ausente' or df1['XML_RJ'][x]=='Sem_Informações':
            faltanteXML[0].append(df1['AGENTE'][x])
            faltanteXML[1].append(df1['STATION NAME'][x])
            faltanteXML[2].append(df1['idName'][x])
dfFaltantesXML = pd.DataFrame (faltanteXML).transpose()
dfFaltantesXML.columns = ['Agente', 'STATION NAME', 'idName']

contagemAusentes = df1['PDC_RJ'][df1['PDC_RJ']=='ausente'].count() 
contagemPresentes = df1['PDC_RJ'][df1['PDC_RJ']=='presente'].count() 
print('Das',len(df1['PDC_RJ']),'PMUs que constam na planilha de Metadados', contagemAusentes, 'não estão configuradas no OpenPDC.')
print('Onde, dentre as', contagemPresentes, 'presentes no OpenPDC', len(dfFaltantesXML['Agente']), 'estão ausentes no XML.\n')
print('Abaixo são sinalizadas as PMUs faltantes para cada análise:')

print('\n#####################################################')
print('Os fluxos presentes na planilha de Metadados, mas ausentes no OpenPDC são: \n')
for x in range(len(fluxosausentes)):
    print(fluxosausentes[x])
print('\n#####################################################')
print('Abaixo as PMUs presentes no OpenPDC, mas ausentes no XML\n')
print(dfFaltantesXML)


#Aqui eu considerei as novas PMUs como faltantes, pois elas não apresentam os idNames preenchidos 
#Informação essencial para a busca das mesmas no XML.
```

```python
# code for displaying multiple images in one figure

#import libraries
import cv2
from matplotlib import pyplot as plt

# create figure
fig = plt.figure(figsize=(20,8))

# setting values to rows and column variables
rows = 2
columns = 2

# reading images
Image1 = cv2.imread('Metadados_PDC_BSB.png')
Image2 = cv2.imread('Metadados_PDC_RJ.png')
Image3 = cv2.imread('Metadados_XML_BSB.png')
Image4 = cv2.imread('Metadados_XML_RJ.png')

# Adds a subplot at the 1st position
fig.add_subplot(rows, columns, 1)
print('Metadados x OpenPDC')

# showing image
plt.imshow(Image1)
plt.axis('off')
plt.title("BSB")

# Adds a subplot at the 2nd position
fig.add_subplot(rows, columns, 2)

# showing image
plt.imshow(Image2)
plt.axis('off')
plt.title("RJ")

# Adds a subplot at the 3rd position
fig.add_subplot(rows, columns, 3)

# showing image
plt.imshow(Image3)
plt.axis('off')
plt.title("BSB")

# Adds a subplot at the 4th position
fig.add_subplot(rows, columns, 4)

# showing image
plt.imshow(Image4)
plt.axis('off')
plt.title("RJ")

```
