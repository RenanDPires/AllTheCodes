```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import re
import os 
path = r"C:\Users\DUARTE\Documents\GitHub\AllTheCodes\Estágio\Checagem de fluxos\Configs\RJ" 
os.chdir(path)
print(os.getcwd())
%matplotlib inline
```

```python
for diretorio, subpastas, arquivos in os.walk(path):
    for arquivo in arquivos:
        print(os.path.join(diretorio, arquivo))
```

```python
fluxos= []
quantidade= []
for j in range(len(arquivos)):
    arquivo = open(arquivos[j], 'r')
    limpo= arquivo.readlines()
    for x in range(len(limpo)):
        if 'stationName' in limpo[x]:
            resultado1=re.split(r'<stationName',limpo[x])
            for l in range(len(resultado1)):
                resultado2=re.split(r'</stationName>',resultado1[l])
                resultado3= re.split(r'>',resultado2[0])
            fluxos.append(resultado3[1])
    quantidade.append(len(fluxos))
    arquivo.close()

quantidade_por_arquivo= []
dados=[[],[],[]]
for q in range(len(quantidade)):
    if (q == 0):
        arquivo_x= quantidade[q]
        quantidade_por_arquivo.append(arquivo_x)
    else:
        arquivo_x= quantidade[q]-quantidade[q-1]
        quantidade_por_arquivo.append(arquivo_x)
    dados[0].append(arquivos[q])
    dados[1].append(arquivo_x)
    dados[2].append(quantidade[q])

df = pd.DataFrame (dados).transpose()
df.columns = ['Arquivo|','PMUs por arquivo|','Total so far|']
df_sort= df.sort_values(by='PMUs por arquivo|', ascending = True)
print (df)
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(15,8))
sns.barplot(x="PMUs por arquivo|", y="Arquivo|", edgecolor="0.1", palette = "flare", data=df_sort)
plt.show()
```
