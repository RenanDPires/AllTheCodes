```python
import re
import os
```

```python
arquivo = open('xml_RJ.txt', 'r')
novo = open("xml.txt", "a")
```

```python
print(arquivo)
```

```python
limpo= arquivo.readlines()
```

```python
fluxos= []
for x in range(len(limpo)):
    if '<idName>' in limpo[x]:
        resultado1=re.split(r'<idName>',limpo[x])
        for l in range(len(resultado1)):
            resultado2=re.split(r'</idName>\n',resultado1[l])
        fluxos.append(resultado2[0])
        novo.write(resultado2[0])
print(fluxos)
novo.close()
```

```python
print(len(fluxos))
```

```python
fluxos.sort()
print(fluxos)
```
