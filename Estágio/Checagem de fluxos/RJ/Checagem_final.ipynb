```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import re
import os 
%matplotlib inline
```

```python
busca= open('busca.txt', 'r').readlines()
excel= open('excel.txt', 'r').readlines()
print(len(busca), len(excel))
```

```python
resultados= [[],[]]
for x in range(len(excel)):
    a= excel[x]

    if a in busca:
        resultado= '1'
    else:
        resultado= '0'
    resultados[0].append(resultado)
print(resultados)
len(resultados[0])
```

```python
excel_trocado= []
for i in range(len(excel)):
    troca= excel[i].replace('/','-')
    excel_trocado.append(troca)
    print(excel[i])
    print(troca)
```

```python
for x in range(len(excel)):
    a= excel_trocado[x]
    if a in busca:
        resultado= '1'
    else:
        resultado= '0'
    resultados[1].append(resultado)
print(resultados)
```

```python
novo = open("final.txt", "a")
para_excel=[]
for n in range(len(resultados[0])):
    if resultados[0][n] == '1' or resultados[1][n] == '1':
        checagem_final= '1\n'
        para_excel.append(checagem_final)
    else:
        checagem_final= '0\n'
        para_excel.append(checagem_final)
    novo.write(checagem_final)
print(para_excel)
novo.close()
```
