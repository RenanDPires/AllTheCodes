# 1- Primeiro, importe as bibliotecas pandas, matplotlib e seaborn, e use a palavra mágica inline para exibir as visualizações no notebook. Depois, carregue os dados, que estão disponibilizados neste link: dados, em um DataFrame. Em seguida, você pode carregar os dados pela URI, ou salvá-los em sua máquina e carregá-los localmente;

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")

shampoo_uri= 'https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro07/oficina/arquivos/dadosoficina07.txt'
shampoo= pd.read_csv(shampoo_uri)
shampoo.head()
```

# 2- Para o segundo e último passo, crie uma visualização que mostre a série temporal das vendas ao longo dos meses. Você pode, por exemplo, se perguntar se existe alguma tendência ou alguma periodicidade, OK?

Primeiramente eu reorganizei os dados de ano e mês da seguinte forma:
Quebrei as informações de ano e mês que estavam separadas por '-';
Em seguida criei colunas no df shampoo com os nomes Ano e Mes para que eu pudesse colocar as informações que antes foram quebradas;
E por último excluí a coluna de Month (a que trazia as informações de mês e ano juntas).

```python
novaColuna = shampoo['Month'].str.split("-", n = 1, expand = True)
shampoo['Ano']= novaColuna[0]
shampoo['Mes']= novaColuna[1]
shampoo.drop(['Month'], axis=1)

plt.figure(figsize=(20,5))
sns.lineplot(x='Mes', y='Sales',hue='Ano', data= shampoo)
plt.show()
```

R= A partir da visualização, observamos que: Com o decorrer dos anos as vendas aumentaram, mesmo tendo quedas (sempre se recuperando); Outra infor,ação é que o 3° ano tem sido o melhor da empresa e único que não tem vendas maiores que no mesmo período do ano anterior (o que acontece nos meses 2, 7 e 11 do ano, se comparado com o ano 1).
