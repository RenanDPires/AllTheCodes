```python
especie= ['Cachorro', 'Gato', 'P√°ssaro']
quantidade= [110, 90, 30]
plt.title('Propor√ß√£o de esp√©cies')
plt.pie(x=quantidade, labels = especie)
plt.style.use('seaborn-pastel')
plt.show()
```

Na linha 4, na fun√ß√£o pie() foi passado, como argumento, √† string '%1.1f%%'. Essa formata√ß√£o indica que as porcentagens devem ser exibidas com uma casa decimal. Se voc√™ quiser exibir com duas casas decimais, por exemplo, pode utilizar '%1.2f%%‚Äô. Partindo dessas informa√ß√µes, saiba que a fun√ß√£o pie() pode receber diversos argumentos para controlar o visual dos gr√°ficos. Ent√£o, para uma lista completa, consulte a documenta√ß√£o da fun√ß√£o e os exemplos neste link: https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.pie.html.

```python
especie= ['Cachorro', 'Gato', 'P√°ssaro']
quantidade= [110, 90, 30]
plt.title('Propor√ß√£o de esp√©cies')
plt.pie(x=quantidade, labels = especie, autopct='%1.3f%%')
plt.figure(figsize=(8,8))
plt.style.use('seaborn-pastel')
plt.show()
```

# Gr√°fico de rosca

```python
races = ['Poodle', 'Golden Retriver', 'Bulldog', 'Pastor Alem√£o', 'Labrador', 'Sem Ra√ßa Definida']
quant_races = [25,15,5,20,25,10]
plt.title('Propor√ß√£o de ra√ßas de c√£es')
plt.pie(x=quant_races, labels = races, autopct='%1.2f%%')
plt.show()
```

O matplotlib n√£o possui uma fun√ß√£o pr√≥pria para criar gr√°ficos de rosca, mas eles podem ser criados fazendo um gr√°fico de pizza e desenhando um c√≠rculo branco no centro. Voc√™ pode conferir como isso √© feito de acordo com o c√≥digo e sua aplica√ß√£o a seguir.

```python
plt.pie(x=quant_races, labels=races, autopct='%1.2f%%',startangle=90, pctdistance=0.85)
my_circle=plt.Circle((0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.title('Propor√ß√£o de ra√ßas de c√£es')
plt.show()
```

# Outra representa√ß√£o para propor√ß√µes

```python
games = pd.read_csv('vgsales.txt')
games.head()
```

```python
generos= games.groupby(by= 'Genre').sum()
generos
```

```python
plt.pie(x=generos['Global_Sales'], labels = generos.index, autopct='%1.2f%%',startangle=90, pctdistance=0.85)
plt.show()
```

```python
plt.pie(x=generos['Global_Sales'], labels = generos.index, autopct='%1.1f%%',startangle=90, pctdistance=0.70)
my_circle=plt.Circle((0,0), 0.3, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```

```python
generosVendas = generos.sort_values(by='Global_Sales',ascending=False)
porcentagens = generosVendas['Global_Sales']*100.0/generos['Global_Sales'].sum()
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(15,5))
plt.bar(generos.index,porcentagens)
plt.xlabel('G√™neros')
plt.ylabel('%')
plt.title('Propor√ß√£o de vendas de jogos por g√™nero')
plt.show()
```
