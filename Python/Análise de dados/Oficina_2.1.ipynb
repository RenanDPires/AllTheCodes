# 1.1- Importar bibliotecas e usar o comando mágico

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# 1.2- Carregar o dataframe

```python
filmes_uri = 'https://raw.githubusercontent.com/emanueles/datavis-course/master/assets/files/observable/movies.json'
filmes = pd.read_json(filmes_uri)
filmes.head()
```

# 2- Orçamento de todos os filmes / maior e menor

Criar novo dataframe com os Orçamentos em ordem decrescente

```python
filmesOrcamento= filmes.sort_values(by='Budget_M', ascending = False)
filmesOrcamento.head()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(15,8))
plt.title('Orçamento dos filmes')
sns.barplot(x="Budget_M", y="Film", edgecolor="0.1", palette = "flare", data=filmesOrcamento)
plt.show()
```

R= Com base no gráfico, o filme com maior orçamento é: Piratas do caribe: O fim do mundo.
Assim como também podemos afirmar sobre o filme com menor orçamento: A saga crepúsculo: Lua Nova.

# 3- Bilheteria de todos os filmes

Criar novo dataframe com as bilheterias em ordem decrescente

```python
filmesBilheteria= filmes.sort_values(by='Worldwide_Gross_M', ascending = False)
filmesBilheteria.head()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(15,8))
plt.title('Bilheteria dos filmes')
sns.barplot(x="Worldwide_Gross_M", y="Film", edgecolor="0.1", palette = "flare", data=filmesBilheteria)
plt.show()
```

R= Com base no gráfico, o filme com maior bilheteria é: Avatar.
Assim como também podemos afirmar sobre o filme com menor Bilheteria: Ratatouille.

# 4- Quantidade de filmes por gênero por ano

```python
sns.set(style="whitegrid")
plt.figure(figsize=(15,8))
plt.title('Gêneros de filmes por ano')
sns.countplot(x="Year", hue="Genre", edgecolor="0.1", data=filmes)
plt.show()
```

R= O ano em que foram feitos menos filmes de ação foi em: Mesma quantidade em 2008 e 2010 (1 filme).
E o ano em que menos fizeram filmes de animação foi em: Mesma quantidade e, 2008 e 2011 (1 filme).

# 5- Média de orçamento de filme por gênero em cada ano

```python
sns.set(style="whitegrid")
plt.figure(figsize=(15,8))
plt.title('Média de orçamento por gênero em cada ano')
sns.barplot(x="Year", y='Budget_M', hue='Genre', ci=None, edgecolor="0.1", data=filmes)
plt.show()
```

R= O ano em que os filmes de ação tiveram maior média orçamento foi: 2007.
Já o ano em que os filmes de animação tiveram maior média de orçamento foi: 2008.
