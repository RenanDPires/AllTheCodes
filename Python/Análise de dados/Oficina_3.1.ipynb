# 1- Importe as bibliotecas pandas, matplotlib e seaborn e use a palavra mágica inline para exibir as visualizações no notebook. Carregue os dados em um DataFrame.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

imoveis_uri= 'https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro05/oficina/arquivos/dadosoficina05.txt'
imoveis= pd.read_csv(imoveis_uri)
imoveis
```

# 2- Crie uma visualização que mostre a distribuição dos valores de aluguel, ‘rent amount (Rs)’, dos imóveis. Qual faixa de valores de aluguel é mais comum?

```python
plt.figure(figsize=(20,8))
plt.style.use('seaborn-whitegrid')
plt.title('Distribuição de valor de aluguel')
plt.hist(imoveis['rent amount (R$)'], bins= 100)
plt.xlabel('Valor de aluguel')
plt.ylabel('Quantidade de imóveis')
plt.show()
```

R= Podemos garantir que os valores mais frequentes estão entre 2000 e 3000. (cada barra equivale a 500 no eixo X)

# 2- Crie uma visualização que mostre a distribuição dos valores de seguro contra incêndio, 'fire insurance (Rs)', dos imóveis. Existem valores incomuns?

```python
plt.figure(figsize=(20,8))
plt.style.use('seaborn-whitegrid')
plt.title('Distribuição de valores de seguro contra incêndio')
plt.hist(imoveis['fire insurance (R$)'], bins= 100)
plt.xlabel('valor do seguro')
plt.ylabel('Quantidade de imóveis')
plt.show()
```

R= Podemos observar que a partir R$ 100 há valores que não são frequentes (barra pequena, uqase invisível)
