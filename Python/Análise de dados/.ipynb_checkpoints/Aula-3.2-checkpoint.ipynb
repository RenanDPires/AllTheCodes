# Grandezas quantitativas correlatas (duas variáveis)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

Gráfico de dispersão entre valor total da conta e gorjeta, visando verificar se há correlação entre elas

```python
plt.figure(figsize=(8,5))
plt.title('Relação entre valor da conta e valor da gorgeta')
sns.scatterplot(x='total_bill', y='tip', data= tips)
plt.show()
```

Gráfico de regressão, para verificar (a partir de uma reta calculada) a relação entre as grandezas

```python
plt.figure(figsize=(8,5))
plt.title('Relação entre valor da conta e valor da gorgeta')
sns.regplot(x='total_bill', y='tip', data= tips)
plt.show()
```

# Grandezas quantitativas e qualitativas correlatas (várias variáveis)

Verificação de tendência por gênero

```python
plt.figure(figsize=(8,5))
plt.title('Relação entre valor da conta e valor da gorjeta pelo gênero')
sns.scatterplot(x='total_bill', y='tip', hue='sex', data= tips)
plt.show()
```

Para verificarmos a regressão de cada gênero devemos usar a função lmplot()

```python
sns.lmplot(x='total_bill', y='tip', hue='sex', data= tips)
plt.show()
```

Verificação de tendência por dia (muitas variáveis, logo temos muitas cores que dificultam a compreensão)

```python
plt.figure(figsize=(8,5))
plt.title('Relação entre valor da conta e valor da gorjeta pelo gênero')
sns.scatterplot(x='total_bill', y='tip', hue='day', data= tips)
plt.show()
```

Neste caso devemos usar a função relplot() que é uma versão mais geral da scatterplot(), mas que é baseada em uma classe do seaborn chamada FacetGrid, que permite criar grades de visualizações. Para isso, basta passar o nome da coluna que se deseja separar em grade para o parâmetro col.

```python
sns.relplot(x='total_bill', y='tip', col='day', data= tips)
plt.show()
```

# Grandezas quantitativas correlatas (várias variáveis)

```python
plt.figure(figsize=(8,5))
plt.title('Relação entre valor da conta, valor da gorjeta e tamanho da mesa')
sns.scatterplot(x='total_bill', y='tip', hue='size', data= tips)
plt.show()
```

Gráfico de bolhas
Uma outra maneira de representar essa variável é através do tamanho dos círculos, de modo que quanto maior o círculo, maior o tamanho da mesa. Para isso, a função scatterplot() possui o parâmetro size. No exemplo a seguir, a coluna ‘size’ será passada ao parâmetro size, para mapear o tamanho das mesas no tamanho dos círculos.

```python
plt.figure(figsize=(8,5))
plt.title('Relação entre valor da conta, valor da gorjeta e tamanho da mesa')
sns.scatterplot(x='total_bill', y='tip', size='size', sizes=(15,200), data= tips)
plt.show()
```

# Todas as grandezas quantitativas e qualitativas correlatas (várias variáveis)

As linhas 1 e 2 configuram o tamanho e título da imagem, como já feito anteriormente. Na linha 3, a função scatterplot() recebeu a coluna ‘total_bill’ para o parâmetro x; a coluna ‘tip’ para o parâmetro y; a coluna ‘size’ para o parâmetro size; a coluna ‘sex’ para o parâmetro hue; e a coluna ‘smoker’ para o parâmetro style. Ou seja, o valor da conta foi mapeado para a posição horizontal, a gorjeta para a posição vertical, o tamanho da mesa para o tamanho do ponto, o sexo do cliente para a cor, e se ele é fumante ou não para a forma. Assim, quase todas as variáveis do conjunto de dados foram representadas em uma única visualização.

```python
plt.figure(figsize=(12,8))
plt.title('Gorjetas')
sns.scatterplot(x="total_bill", y="tip", size = 'size', hue = 'sex', style = 'smoker', sizes=(15, 200), data=tips)
plt.show()
```
