# Confecção de gráficos de linha

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

```python
quarto_uri = 'https://raw.githubusercontent.com/cabrau/curso-analise-visualizacao-dados-python/master/room_data.txt'
quarto = pd.read_csv(quarto_uri)
quarto.head()
```

Gráfico de dispersão
Converter a coluna de datas de string para data com o método to_datetime()

```python
quarto['date'] = pd.to_datetime(quarto['date'])
plt.figure(figsize=(10,5))
plt.title('Variação da temperatura ao longo do tempo')
sns.scatterplot(x="date", y="Temperature", data=quarto)
plt.show()
```

```python
quarto['date'] = pd.to_datetime(quarto['date'])
plt.figure(figsize=(10,5))
plt.title('Variação da temperatura ao longo do tempo')
sns.lineplot(x="date", y="Temperature", data=quarto)
plt.show()
```

```python
hiv_uri = 'https://raw.githubusercontent.com/cabrau/curso-analise-visualizacao-dados-python/master/hiv_data.txt'
hiv = pd.read_csv(hiv_uri)
hiv.head()
```

Como são vários valores para um mesmo X, a função lineplot() cria a linha a partir da média e gera uma sombra, que vem a ser a área de confiança. Ou seja, área provável de acharmos um valor de y.

```python
plt.figure(figsize=(10,5))
plt.title('Variação da taxa média de HIV global')
sns.lineplot(x="date", y="HIV Rate", data=hiv)
plt.show()
```

Aqui usamos o ci= None para não ser mostrada a área de confiança

```python
plt.figure(figsize=(10,5))
plt.title('Variação da taxa média de HIV para níveis de renda')
sns.lineplot(x="date", y="HIV Rate", hue = 'Income Level', ci = None, data=hiv)
plt.show()
```

Aqui usamos o parâmetro legend= False para não mostrar a legenda (algo que é apenas para caráter didático)

```python
plt.figure(figsize=(10,5))
plt.title('Variação da taxa de HIV para cada país')
sns.lineplot(x="date", y="HIV Rate", hue = 'country', legend = False, data=hiv)
plt.show()
```
