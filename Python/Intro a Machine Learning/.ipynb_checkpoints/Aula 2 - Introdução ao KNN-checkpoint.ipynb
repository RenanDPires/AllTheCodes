# Distância Euclidiana

Como se fosse uma distância modular, ou uma distância absoluta entre pontos.

Abaixo, distância não euclidiana para apenas uma dimensão:

```python
2-5
```

Abaixo, distância euclidiana para apenas uma dimensão:

```python
((2-5)**2)**(1/2)
```

Para mais de uma dimensão:

```python
a= [5,0.75]
b= [2,0.50]
```

```python
resultado= []
for x in range(len(a)):
    c= ((a[x]-b[x])**2)**0.5
    resultado.append(c)
print(resultado)
```

Tamanho do vetor

```python
vetor= (resultado[0]**2+resultado[1]**2)**0.5
print(vetor)
```
